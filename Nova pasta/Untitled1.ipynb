```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading dataset

```python
campos1 = pd.read_csv('dado_campos1.csv', sep='\s+')
campos1['DEPTH'] = campos1['DEPTH']*0.3048
campos1 = campos1[campos1['ILD'] < 500]
campos1.corr()
```

```python
campos3 = pd.read_csv('dado_campos3.csv', sep='\s+')
campos3['DEPTH'] = campos3['DEPTH']*0.3048
campos3.corr()
campos4 = pd.read_csv('dado_campos4.csv', sep='\s+')
campos4['DEPTH'] = campos4['DEPTH']*0.3048
campos4.corr()
```

```python
campos_concat = pd.concat([campos1, campos3], axis=0)
campos_concat.corr()
```

```python
X_campos_ref = campos_concat.iloc[:,[0,1,3,4]]
y_campos_ref = campos_concat.iloc[:,5]
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_campos_ref_sclnd = scaler.fit_transform(X_campos_ref)
```

```python
X_campos_ref
```

```python
svr = SVR().fit(X_campos_ref_sclnd, y_campos_ref)
```

```python
X_campos4 = campos4.iloc[:,[0,1,3,4]]
X_campos4_sclnd = scaler.transform(X_campos4)
y_campos4 = campos4.iloc[:,5]
```

```python
X_campos4
```

```python
y_campos4_pred = svr.predict(X_campos4_sclnd)
```

```python
from sklearn.metrics import r2_score as r2
```

```python
r2(y_campos4,y_campos4_pred)
```
